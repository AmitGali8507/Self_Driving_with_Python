## convert from RGB to HSV

```python
image = cv2.imread("Self_Driving_Car.jpg")
cv2.imshow("Self_Driving_Car", image)

hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
cv2.imshow("HSV Image", hsv_image)

cv2.waitKey()
cv2.destroyAllWindows()
```

```python
plt.imshow(hsv_image[:,:,0])
plt.title("HUE Channel")
```

```python
plt.imshow(hsv_image[:,:,1])
plt.title("Saturation Channel")
```

```python
plt.imshow(hsv_image[:,:,2])
plt.title("Value Channel")
```

```python
plt.imshow(hsv_image[:,:,:])
plt.title("Full HSV")
# which will be different to cv2
```

## split and merge R,G,B channels

```python
image = cv2.imread("Self_Driving_Car.jpg")
B,G,R = cv2.split(image)
B.shape
```

```python
R.shape
```

```python
G.shape
```

```python
cv2.imshow("Blue Channel", B)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
cv2.imshow("Blue Channel", G)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
cv2.imshow("Blue Channel", R)
cv2.waitKey()
cv2.destroyAllWindows()
```

## lets create our own 3D image out of the blue channel

```python
zeros = np.zeros(image.shape[:2], dtype = "uint8")


cv2.imshow("Blue Channel", cv2.merge([B, zeros, zeros]))
cv2.waitKey()
cv2.destroyAllWindows()
```

## lets merge RBG channels to create our original image

```python
image_merged = cv2.merge([B, G, R])

cv2.imshow("Merged Image", image_merged)
cv2.waitKey()
cv2.destroyAllWindows()
```

## lets merge again while adding more green

```python
image_merged = cv2.merge([B, G + 100, R])

cv2.imshow("Merged Image", image_merged)
cv2.waitKey()
cv2.destroyAllWindows()
```
