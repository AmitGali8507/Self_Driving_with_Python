## CODE TO PERFORM REGION OF INTEREST (ROI) SELECTION

```python
import cv2
import numpy as np
import matplotlib.image as mpimg
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
image_color = cv2.imread("image_lane_c.jpg")
```

```python
cv2.imshow('Original Image', image_color)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
height, width = image_color.shape[:2]
```

## convert ot grayscale

```python
image_gray = cv2.cvtColor(image_color, cv2.COLOR_BGR2GRAY)
plt.imshow(image_gray,cmap="gray")
```

```python
image_gray.shape
```

## region of interest masking

```python
ROI = np.array([[(0,height),(400,350),(550,350),(width,height)]], dtype = np.int32)
```

```python
blank = np.zeros_like(image_gray)
```

```python
mask = cv2.fillPoly(blank, ROI, 255)
```

```python
masked_image = cv2.bitwise_and(image_gray, mask)
```

```python
masked_image.shape
```

```python
plt.imshow(masked_image,cmap="gray")
```
