# multi lyer perceptrons

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn import datasets
```

```python
noisy_moons = datasets.make_moons(n_samples=1500, noise=.05)
```

```python
X, y = noisy_moons
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
```

```python
import tensorflow.keras
```

```python
from keras.models import Sequential
```

```python
from keras.layers import Dense
```

```python
from sklearn.datasets import make_blobs
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
scaler = MinMaxScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train_scaled = scaler.transform(X_train)
```

```python
scaler.fit(X_test)
```

```python
X_test_scaled = scaler.transform(X_test)
```

```python
plt.scatter(X_train_scaled[:,0], X_train_scaled[:,1], c=y_train)
```

## fit the final model

```python
model =Sequential()
```

```python
model.add(Dense(1, input_dim=2, activation="sigmoid"))
```

```python
model.compile(loss="binary_crossentropy", optimizer="adam")
```

```python
model.fit(X_train_scaled, y_train, epochs=200, verbose=2)
```

## try with more neurons

```python
model = Sequential()
model.add(Dense(20, input_dim=2, activation="relu"))
model.add(Dense(20, activation="relu"))
model.add(Dense(20, activation="relu"))
model.add(Dense(20, activation="relu"))
model.add(Dense(1, activation="sigmoid"))
model.compile(loss="binary_crossentropy", optimizer="adam")
model.fit(X_train_scaled, y_train, epochs=200, verbose=2)
```

```python
model.summary()
```

```python
y_predict = (model.predict(X_test_scaled) > 0.5).astype("int32")
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
cm = confusion_matrix(y_test, y_predict)
```

```python
sns.heatmap(cm, annot=True, fmt="d")
```

```python
print(classification_report(y_test,y_predict))
```

```python
# Visualising the Training set results
from matplotlib.colors import ListedColormap
X_grid, y_grid = X_train_scaled, y_train
X1, X2 = np.meshgrid(np.arange(start = X_grid[:, 0].min() - 1, stop = X_grid[:, 0].max() + 1, step = 0.01),
                     np.arange(start = X_grid[:, 1].min() - 1, stop = X_grid[:, 1].max() + 1, step = 0.01))
plt.contourf(X1, X2, model.predict(np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_grid)):
    plt.scatter(X_grid[y_grid == j, 0], X_grid[y_grid == j, 1],
                c = ListedColormap(('red', 'green'))(i), label = j)
    
plt.title('Training dataset')
plt.xlabel('X1')
plt.ylabel('X2')
plt.legend()
plt.show()
```
